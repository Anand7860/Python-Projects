```python
#OOPS Concept (Inheritance )
#Parent Class
class Employee(object):
    def __init__(self,Name,Department):
        self.Name = Name
        self.Department = Department
        
    def display(self):
        print(self.Name)
        print(self.Department)
#Child Class
class Matching(Employee):
    def __init__(self, Name, Department, Monthlysalary):
        self.Monthlysalary = Monthlysalary
        
        Employee.__init__(self, Name, Department)
        
        
    def details(self):
        print("My Name is: {}".format(self.Name))
        print("Department: {}".format(self.Department))
        print("Monthlysalary: {}".format(self.Monthlysalary))
    
        
a = Matching('A1','Sales', 20000)
b = Matching('B1','Maintainence',20000)
c = Matching('C1','IT',80000)
d = Matching('D1','Research & Development',30000)
e = Matching('E1','Human Resources',25000)
  
# calling a function of the class Person using
# its instance
a.display()
a.details()
b.display()
b.details()
c.display()
c.details()
d.display()
d.details()
e.display()
e.details()

    
```
